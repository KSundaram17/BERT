```python
# make a request to google to get synthesis  
t1 = gtts.gTTS("This classic fable (story) tells the story of a very slow tortoise (another word for turtle) and a speedy hare (another word for rabbit). The tortoise challenges the hare to a race. The hare laughs at the idea that a tortoise could run faster than him, but when the two actually race, the results are surprising.") 
```

```python
# save the audio file  
t1.save("longtext.mp3")
```
